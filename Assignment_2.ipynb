```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

Task-1 Download the dataset

 House Price India dataset is downloaded.

Task-2 Load the dataset

```python
df=pd.read_csv('/content/House Price India.csv')
df
```

task-3 Perform the Below Visualizations. ￿ Univariate Analysis ￿ Bi - Variate Analysis ￿ Multivariate
Analysis

```python
# Univariate Analysis
#(Analysis on a single feature 'Built Year')

sns.distplot(df['Built Year'])
```

```python
# Univariate Analysis
# Generate histograms for numerical variables

df.hist(figsize=(12, 10))
plt.tight_layout()
plt.show()
```

```python
#univariate analysis
#box plot
sns.boxplot(df["living area"])
plt.title('living area')
```

```python
#uni-variate analysis
#barplot

sns.barplot(x=df["number of bedrooms"].value_counts().index,y=df["number of bedrooms"].value_counts())
```

```python
#univariate analysis
#piechart
plt.pie(df["number of floors"].value_counts(),[0.1,0,0.1,0.1,0.1,0.1],autopct='%1.1f%%')
plt.title('number of floors')
plt.show()
```

```python
# Bivariate Analysis
#(Comparision between 'number of floors' feature and 'Price')

sns.jointplot(x=df['number of floors'],y='Price',data=df)
```

```python
# Bivariate Analysis
#(Comparision between 'lot_area' feature and 'Price')

sns.jointplot(x=df['Distance from the airport'],y='Price',data=df)
```

```python
# bivariate analysis
#scatterplot
sns.scatterplot(x=df["living area"],y=df["Price"])
```

```python
#bivariate analysis
sns.lineplot(x=df["number of bathrooms"],y=df["Price"])
```

```python
# Multivariate Analysis
# Generate scatter plots for numerical variables
numerical_vars = ['living area', 'lot area', 'Area of the house(excluding basement)', 'Price']
sns.pairplot(data=df, vars=numerical_vars)
plt.show()
```

```python
# Multivariate Analysis
# Generate correlation matrix and heatmap

corr_matrix = df.corr()
plt.figure(figsize=(12, 10))
sns.heatmap(corr_matrix, annot=True)
plt.title('Correlation Matrix')
plt.show()
```
